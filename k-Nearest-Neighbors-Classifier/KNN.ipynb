```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    "X1": [1,3,2,2,1,3,2,3,2,1],
    "X2": [1,3,3,2,1,4,2,4,3,2],
    "X3": [1,2,1,1,1,4,2,3,2,1],
    "Y":  [1,2,2,1,1,3,1,3,2,1]
})


train = df.iloc[:8]
test  = df.iloc[8:]

k = 6


def knn(train_df, test_row):

    distances = []
    print(f'{test_row } \n')
    for i in range(len(train_df)):

        x_train = train_df.iloc[i][["X1","X2","X3"]].values
        x_test  = test_row[["X1","X2","X3"]].values

        d = np.sqrt(np.sum((x_train - x_test) ** 2))

        label = train_df.iloc[i]["Y"]
        print(f'{i+1} -> d = {d}, y = {label}')
        distances.append((d, label))

    distances.sort()


    nearest = [distances[i][1] for i in range(k)]

    count1 = 0
    count2 = 0
    count3 = 0

    for label in nearest:
        if label == 1:
            count1 += 1
        elif label == 2:
            count2 += 1
        else:
            count3 += 1


    if count1 >= count2 and count1 >= count3:
        return 1
    elif count2 >= count1 and count2 >= count3:
        return 2
    else:
        return 3




```

```python

correct = 0
s = 0
for i in range(len(test)):

    g = knn(train, test.iloc[i])
    true_label = test.iloc[i]["Y"]
    s += np.power(true_label - g, 2)
    print(" \n irakan:", true_label, "kanxatesel:", g, '\n')

    if g == true_label:
        correct += 1

```

```python
ep = np.sqrt(s/len(test))
print("epsilion:", ep)
print("true", correct/len(test))
```
