```python
import numpy as np
import pandas as pd

points = np.array([
    [2, 6],   # A
    [3, 4],   # B
    [4, 7],   # C
    [7, 3],   # D
    [8, 5],   # E
    [6, 8]    # F
])

names = ["A", "B", "C", "D", "E", "F"]

c1 = np.array([2, 6], dtype=float)
c2 = np.array([8, 5], dtype=float)

def distance(p, c):
    return np.sqrt(np.sum((p - c) ** 2))

rows = []

cluster1 = []
cluster2 = []

for i in range(len(points)):
    p = points[i]

    d1 = distance(p, c1)
    d2 = distance(p, c2)

    if d1 < d2:
        group = 1
        cluster1.append(p)
    else:
        group = 2
        cluster2.append(p)

    rows.append([
        names[i],
        p[0],
        p[1],
        round(d1, 2),
        round(d2, 2),
        group
    ])

df = pd.DataFrame(rows, columns=[
    "Point", "X", "Y", "Dist to C1", "Dist to C2", "Cluster"
])

print("===== FIRST ITERATION =====")
print(df)

cluster1 = np.array(cluster1)
cluster2 = np.array(cluster2)

new_c1 = np.mean(cluster1, axis=0)
new_c2 = np.mean(cluster2, axis=0)

print("\n===== NEW CENTROIDS =====")
print("C1 =", new_c1)
print("C2 =", new_c2)
```
