```python
import numpy as np

X = np.array([
    [18,6,90],
    [20,5,85],
    [22,4,88],
    [21,2,60],
    [24,1,55],
    [19,2,58],
    [18,6,80],
    [21,4,50]
])

y = np.array([
    0,0,0,1,1,1,0,1
])

test = np.array([21,4,87])

def gaussian(x, mean, var):
    return (1 / np.sqrt(2*np.pi*var)) * np.exp(-((x-mean)**2)/(2*var))

classes = np.unique(y)

for cls in classes:
    data = X[y == cls]

    mean = np.mean(data, axis=0)
    var = np.var(data, axis=0)

    prior = len(data)/len(X)

    score = prior

    for i in range(len(test)):
        score *= gaussian(test[i], mean[i], var[i])

    print(cls)
    print("mean =", mean)
    print("var =", var)
    print("score =", score)
    print()
```

```python
import numpy as np
import pandas as pd

points = np.array([
    [2, 6],   # A
    [3, 4],   # B
    [4, 7],   # C
    [7, 3],   # D
    [8, 5],   # E
    [6, 8]    # F
])

names = ["A", "B", "C", "D", "E", "F"]

c1 = np.array([2, 6], dtype=float)
c2 = np.array([8, 5], dtype=float)

def distance(p, c):
    return np.sqrt(np.sum((p - c) ** 2))

rows = []

cluster1 = []
cluster2 = []

for i in range(len(points)):
    p = points[i]

    d1 = distance(p, c1)
    d2 = distance(p, c2)

    if d1 < d2:
        group = 1
        cluster1.append(p)
    else:
        group = 2
        cluster2.append(p)

    rows.append([
        names[i],
        p[0],
        p[1],
        round(d1, 2),
        round(d2, 2),
        group
    ])

df = pd.DataFrame(rows, columns=[
    "Point", "X", "Y", "Dist to C1", "Dist to C2", "Cluster"
])

print("===== FIRST ITERATION =====")
print(df)

cluster1 = np.array(cluster1)
cluster2 = np.array(cluster2)

new_c1 = np.mean(cluster1, axis=0)
new_c2 = np.mean(cluster2, axis=0)

print("\n===== NEW CENTROIDS =====")
print("C1 =", new_c1)
print("C2 =", new_c2)
```
