```python
import math

def sigmoid(x):
    return 1 / (1 + math.exp(-x))


x1 = 1
x2 = 0.5
x3 = 0.1

y_true = 1

alpha = 0.1
ep = 0.05


V01, V02 = 0.1, 0.1
V11, V12 = 0.3, 0.2
V21, V22 = 0.2, 0.2
V31, V32 = 0.3, 0.3

W0 = 0.1
W1 = 0.2
W2 = -0.1

epoch = 0
max_epoch = 10000

while epoch < max_epoch:

    epoch += 1

    print(f"\n{'#'*70}")
    print(f"# ԴԱՐԱՇՐՋԱՆ {epoch}")
    print(f"{'#'*70}")


    zin1 = V01 + V11*x1 + V21*x2 + V31*x3
    zin2 = V02 + V12*x1 + V22*x2 + V32*x3

    z1 = sigmoid(zin1)
    z2 = sigmoid(zin2)

    yin = W0 + W1*z1 + W2*z2
    y_pred = sigmoid(yin)

    print("\n── Թաքնված շերտ ─────────────────────────────")

    print(f"zin1 = {V01} + {V11}*{x1} + {V21}*{x2} + {V31}*{x3}")
    print(f"     = {zin1}")

    print(f"zin2 = {V02} + {V12}*{x1} + {V22}*{x2} + {V32}*{x3}")
    print(f"     = {zin2}")

    print(f"z1 = sigmoid({zin1}) = {z1}")
    print(f"z2 = sigmoid({zin2}) = {z2}")

    print("\n── Ելքային շերտ ─────────────────────────────")

    print(f"yin = {W0} + {W1}*{z1} + ({W2})*{z2}")
    print(f"    = {yin}")

    print(f"y_hat = sigmoid({yin}) = {y_pred}")



    error = y_true - y_pred
    mse = 0.5 * (error ** 2)

    print("\n── Սխալ ─────────────────────────────────────")

    print(f"error = {y_true} - {y_pred} = {error}")
    print(f"MSE = {mse}")


    if mse < ep:
        print("\n Ցանցը հասավ պահանջվող սխալին")
        break


    sigma_k = error * y_pred * (1 - y_pred)

    print("\n── sigma_k ──────────────────────────────────")
    print(f"sigma_k = {sigma_k}")


    dW0 = alpha * sigma_k
    dW1 = alpha * sigma_k * z1
    dW2 = alpha * sigma_k * z2

    print("\n── ΔW ───────────────────────────────────────")

    print(f"dW0 = {dW0}")
    print(f"dW1 = {dW1}")
    print(f"dW2 = {dW2}")


    sigma_in1 = sigma_k * W1
    sigma_in2 = sigma_k * W2

    sigma_1 = sigma_in1 * z1 * (1 - z1)
    sigma_2 = sigma_in2 * z2 * (1 - z2)

    print("\n── Թաքնված շերտի sigma ─────────────────────")

    print(f"sigma_1 = {sigma_1}")
    print(f"sigma_2 = {sigma_2}")



    dV01 = alpha * sigma_1
    dV11 = alpha * sigma_1 * x1
    dV21 = alpha * sigma_1 * x2
    dV31 = alpha * sigma_1 * x3

    dV02 = alpha * sigma_2
    dV12 = alpha * sigma_2 * x1
    dV22 = alpha * sigma_2 * x2
    dV32 = alpha * sigma_2 * x3

    print("\n── ΔV ───────────────────────────────────────")

    print(f"dV01 = {dV01}")
    print(f"dV11 = {dV11}")
    print(f"dV21 = {dV21}")
    print(f"dV31 = {dV31}")

    print()

    print(f"dV02 = {dV02}")
    print(f"dV12 = {dV12}")
    print(f"dV22 = {dV22}")
    print(f"dV32 = {dV32}")



    W0 += dW0
    W1 += dW1
    W2 += dW2

    V01 += dV01
    V11 += dV11
    V21 += dV21
    V31 += dV31

    V02 += dV02
    V12 += dV12
    V22 += dV22
    V32 += dV32

    print("\n── Նոր W ────────────────────────────────────")

    print(f"W0 = {W0}")
    print(f"W1 = {W1}")
    print(f"W2 = {W2}")

    print("\n── Նոր V ────────────────────────────────────")

    print(f"V01 = {V01}")
    print(f"V11 = {V11}")
    print(f"V21 = {V21}")
    print(f"V31 = {V31}")

    print()

    print(f"V02 = {V02}")
    print(f"V12 = {V12}")
    print(f"V22 = {V22}")
    print(f"V32 = {V32}")


print(f"\n{'='*70}")
print("ՎԵՐՋՆԱԿԱՆ ԿՇԻՌՆԵՐ")
print(f"{'='*70}")

print(f"W0 = {W0}")
print(f"W1 = {W1}")
print(f"W2 = {W2}")

print()

print(f"V01 = {V01}")
print(f"V11 = {V11}")
print(f"V21 = {V21}")
print(f"V31 = {V31}")

print()

print(f"V02 = {V02}")
print(f"V12 = {V12}")
print(f"V22 = {V22}")
print(f"V32 = {V32}")
```
